## 生日问题
  

```python
# 有 N 个人不在同一天过生日的概率
N = 365
def partial_factorial_iterative(x, n):
    result = 1
    for i in range(x, x - n, -1):
        result *= i
    return result
```

```python
NS =  range(1, N+1)
probabilitys = [partial_factorial_iterative(365, n)/(365**n) for n in NS]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))
plt.scatter(NS,  probabilitys, linewidths=0.01)
plt.yticks([i/10 for i in range(11)])
plt.xticks([i for i in range(0, 366, 10)])
plt.xlabel("N")
plt.ylabel("Probability")
plt.grid()
plt.show()
```
